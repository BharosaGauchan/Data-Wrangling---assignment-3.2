## 4. Add a column to the dataframe indicating whether or not the Earthquake happened in a country or US stat that si on the Ring of Fire.

```python
# rof list
rof_places = ["Bolivia", "Chile", "Ecuador", "Peru", "Costa Rica", "Guatemala", "Mexico", "Japan", "Philippines", "Indonesia", "New Zealand"
               , "Antarctic", "Canada", "Fiji", "Alaska", "Washington", "California", "Russia", "Taiwan", "Tonga", "Kermadec Islands"]

# add rof column
df["ring_of_fire"] = df["parsed_place"].isin(rof_places)
df
```
